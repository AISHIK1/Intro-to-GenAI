```python
from youtube_transcript_api import YouTubeTranscriptApi,TranscriptsDisabled
from langchain_core.runnables import RunnableParallel,RunnablePassthrough,RunnableSerializable,RunnableLambda
from langchain_core.prompts import PromptTemplate
from langchain_community.vectorstores import FAISS
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_huggingface import HuggingFaceEmbeddings,HuggingFaceEndpoint,ChatHuggingFace
from dotenv import load_dotenv
from langchain_core.output_parsers import StrOutputParser
```

```python
from youtube_transcript_api import YouTubeTranscriptApi
from youtube_transcript_api._errors import TranscriptsDisabled

video_id = 'iOdFUJiB0Zc'

try:
    api = YouTubeTranscriptApi()

    transcript_list = api.fetch(video_id)

    transcript = " ".join(
        chunk.text for chunk in transcript_list
    )

    print(transcript)

except TranscriptsDisabled:
    print('Video not found or transcripts are disabled.')

except Exception as e:
    print(f"An unexpected error occurred: {e}")
```

```python
transcript_list
```

```python
splitter=RecursiveCharacterTextSplitter(
    chunk_size=1000,
    chunk_overlap=200)

chunks=splitter.create_documents([transcript])
```

```python
chunks
```

```python
embedding=HuggingFaceEmbeddings(
    model_name="BAAI/bge-base-en-v1.5"
)

vector_store=FAISS.from_documents(chunks,embedding)
```

```python
vector_store.index_to_docstore_id
```

```python
retreiver=vector_store.as_retriever(search_type='similarity',kwargs={'k':5})
```

```python
llm=HuggingFaceEndpoint(
    repo_id='Qwen/Qwen2.5-7B-Instruct',
    task='text-generation',
    temperature=0.5
)

model=ChatHuggingFace(llm=llm)
```

```python
template=PromptTemplate(
    template="""You are helpful aI assistant. Provide the answer to the quetion from the context given. If you dont have the answer to the question from the provided content just ask I dont know
    \n
    {context}
    \n
    \n
    {question}""",
    input_variables=['context','question']
)
```

```python
query='is the topic of AI and olama discussed in this video? if yes then what was discussed'
docs=retreiver.invoke(query)
docs
```

```python
context_text="\n".join([doc.page_content for doc in docs])
```

```python
prompt=template.invoke({
    'context':context_text,
    'question':query
})
```

```python
result=model.invoke(prompt)
```

```python
result.content
```

```python
# now using chanins to form the whole strcuture

def format_word(text):
  return "\n\n".join([doc.page_content for doc in docs])
```

```python
parallel_chain=RunnableParallel({
    'context':retreiver | RunnableLambda(format_word),
    'question':RunnablePassthrough()
})
```

```python
parser=StrOutputParser()
```

```python
chain=parallel_chain | template | model | parser
```

```python
result=chain.invoke('what has been talked about AI and Olama list out in points ')
```
